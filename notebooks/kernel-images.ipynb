## Default docker image

The default image comes with the following:
1. Jupyterlab internal libraries and applications
2. Actual ytsaurus-client and ytsaurus-yson
3. Spark tools: python spark libraries and openjdk
4. numpy
5. pandas
6. matpotlib


# How to build custom Jupyter Kernel image

## Create Dockerfile

Let's assume we need to build a custom image for Jupyter with AWS CLI. In that case, we can take the default image and build a new image based on it.

```docker
> cat ./Dockerfile

FROM ghcr.io/tractoai/notebook-kernel-default:2024-12-17-18-49-00-ceb64d083

RUN curl "https://awscli.amazonaws.com/awscli-exe-linux-x86_64.zip" -o "awscliv2.zip" \
    && unzip awscliv2.zip \
    && ./aws/install \
    && rm -rf ./aws ./awscliv2.zip

RUN aws --version
```

The latest version of the default kernel you can see on the kernel's creation page or in this notebook.

## Build and push image

Assuming we have a Docker registry `cr.example.com`, we can build a new image and push it to the registry. It's important to set platform to `linux/amd64`.

```bash
docker build -t cr.example.com/images/kernel_with_aws:v1 --platform linux/amd64 .
docker push cr.example.com/images/kernel_with_aws:v1
```

## Use custom image in new kernel

Create a new kernel on the [page](/playground/kernels), specify the image as `cr.example.com/images/kernel_with_aws:v1`.

## Authorization in registry

Some Docker registries require specifying a username and password to gain access to the image. To use the required username and password for authentication with the Docker registry, you need to specify them in a `docker_auth` secret in `yson` format.

## Lightweight base image

In some cases, using the default image is not suitable, for example, due to library version conflicts. Under these circumstances, you can start with a lightweight base image and install all the required libraries yourself.

The lightweight image comes with the following libraries:
1. Jupyterlab internal libraries and applications.
2. Actual `ytsaurus-client` and `ytsaurus-yson`.

Dockerfile example:

```docker
FROM docker pull ghcr.io/tractoai/notebook-kernel-base:2024-12-27-12-56-33-307b3bc68

RUN pip install pandas==2.1.0

# lightweight image run in a root context
USER 1000
```
