### Tracto Wizard
Tracto Wizard is a new tool to build Docker images directly from the UI and upload them to the [Internal Registry](./docker-registry.ipynb).
Now it can be useful for creation new Jupyter kernels and building images for operations.

#### What the Wizard can:
* Build and upload docker images to internal egistry without any additional tokens/passwords/etc
* Show crosslinks to [internal registry](./docker-registry.ipynb) and build history
* Attach cypress files to a build

#### How to try:
* Click on create object->Blueprint (see screenshot)

* Set image name (it's a cypress path without //) and write Dockerfile content
* Click on "start new build"
* See a link to a builder's operation
